```python
requests.get("https://ocw.mit.edu/courses/electrical-engineering-and-computer-science/6-006-introduction-to-algorithms-spring-2020/lecture-videos/#")
```

```python
response=requests.get("https://ocw.mit.edu/courses/electrical-engineering-and-computer-science/6-006-introduction-to-algorithms-spring-2020/lecture-videos/#")
```

```python
soup= BeautifulSoup(response.text)
soup
```

```python
soup.find_all("a",attrs={"class":"bullet medialink"})
```

```python
videos=soup.find_all("a",attrs={"class":"bullet medialink"})
videos
```

```python
videos[0]
```

```python
videos[0].text
```

```python
videos[0].attrs
```

```python
videos[0].attrs["href"]
```

```python
for e in videos:
    print(e.text,"|","https://ocw.mit.edu/"+e.attrs["href"])
```

```python
import csv
```

```python
with open ("introduction to algorithms.csv","w",newline="") as f:
    writer= csv.DictWriter(f,fieldnames=["video name","URL"])
    writer.writeheader()
    for e in videos:
        video_name_=e.text
        url_="https://ocw.mit.edu"+e.attrs["href"]
        writer.writerow(
            {
              "video name": video_name_,
              "URL": url_
              
            }
        )
```
